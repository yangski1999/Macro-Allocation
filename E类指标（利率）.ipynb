```python

def get_line_plot(df, var, var_name=None, name=None, color='red', if_percent=False, min_=None, \
                  if_datazoom=False, datazoom_start=50, datazoom_end=100):
    """
    画单Y轴坐标折线图,代码可复用
    """
    if if_percent:
        foramt = "{value} %"
    else:
        foramt = ""

    if min_ == 'min':
        min_ = df[var].min()

    if var_name is None:
        var_name = var

    line = (
        Line()
        .add_xaxis(list(df.index))
        .add_yaxis(
            series_name=var,
            y_axis=[df[x].tolist() for x in var],
            yaxis_index=0,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=2, ), color=color)
        
    )
    
    if if_datazoom:
        line.set_global_opts(
            title_opts=opts.TitleOpts(title=name),
            legend_opts=opts.LegendOpts(pos_left="center"),
            yaxis_opts=opts.AxisOpts(axistick_opts=opts.AxisTickOpts(is_show=True),
                                     axisline_opts=opts.AxisLineOpts(is_show=True), is_show=True,
                                     min_=min_, name=var_name,
                                     axislabel_opts=opts.LabelOpts(formatter=foramt)),
            tooltip_opts=opts.TooltipOpts(is_show=True, axis_pointer_type="cross", trigger="axis"),
            datazoom_opts=opts.DataZoomOpts(orient='horizontal', range_start=datazoom_start, range_end=datazoom_end))
    else:
        line.set_global_opts(
            title_opts=opts.TitleOpts(title=name),
            
            legend_opts=opts.LegendOpts(pos_left="center"),
            yaxis_opts=opts.AxisOpts(axistick_opts=opts.AxisTickOpts(is_show=True),
                                     axisline_opts=opts.AxisLineOpts(is_show=True), is_show=True,
                                     min_=min_, name=var_name,
                                     axislabel_opts=opts.LabelOpts(formatter=foramt)),
            tooltip_opts=opts.TooltipOpts(is_show=True, axis_pointer_type="cross", trigger="axis"))
   
    return line
```

```python
temp=w.edb("M0041964,M9003295", "2016-11-28", "2023-11-28","Fill=Previous")
data=pd.DataFrame(temp.Data,columns=temp.Times,index=temp.Codes)
data.index=["全部工业增加值：同比","规模以上工业增加值：同比"]
data=data.T
def get_bar_plot(data,var_name=None,name=None,rotate=0,if_percent=False,color = 'skyblue',min_ = None):

    bar = (
        Bar()
        .add_xaxis(list(data.index))
        .add_yaxis(data.columns[0],round(data[data.columns[0]],2).tolist(), color=color )  # 自定义颜色
        .add_yaxis(data.columns[1],round(data[data.columns[1]],2).tolist(), color='purple')
        .set_global_opts(
            title_opts={"text": name},
            xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=0)),  ###设置旋转角度问题
            yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(),min_=min_),
            )
    )    
    return bar
m=get_bar_plot(data)
m.render("indus_1.html")
m.render_notebook()
```
